# Minimum Cost Spanning Tree
Problem Description:
A spanning tree of a graph can be defined as a graph with minimal set of edges that connect all vertices. A minimum spanning tree of a graph is a spanning tree of the graph with least weight (where the weight is computed by adding the weights of all the edges in the spanning tree). In general, a graph can have multiple minimum spanning trees. The problem is to find a minimum spanning tree of a graph.

This implementation uses a dictionary to represent the weighted graph, where the vertices are keys, and the values are dictionaries containing their adjacent vertices and their weights.

Please note that this implementation assumes that the input graph is connected and weighted.

First, let's create a simple graph to demonstrate Prim's algorithm:

```python
graph = {
    "A": {"B": 10, "C": 15, "D": 20},
    "B": {"A": 10, "C": 12},
    "C": {"A": 15, "B": 12, "D": 18},
    "D": {"A": 20, "C": 18},
}
```

Now, let's implement Prim's algorithm:

```python
import heapq

def prim_algorithm(graph):
    min_tree = []  # Stores the edges of the MST
    visited = set() # Stores visited vertices
    min_weight = 0
    # Start from an arbitrary vertex (e.g., the first one in the graph)
    start_vertex = list(graph.keys())[0]
    visited.add(start_vertex)

    # Priority queue to store (weight, u, v) edges
    # The priority queue will always give the minimum weight edge
    edges = []
    for neighbor, weight in graph[start_vertex].items():
        heapq.heappush(edges, (weight, start_vertex, neighbor))

    while edges and len(visited) < len(graph):
        weight, u, v = heapq.heappop(edges)

        if v not in visited:
            visited.add(v)
            min_tree.append((u, v, weight))
            min_weight += weight

            # Add all edges connected to the newly added vertex 'v' to the priority queue
            for neighbor, new_weight in graph[v].items():
                if neighbor not in visited:
                    heapq.heappush(edges, (new_weight, v, neighbor))

    return min_tree, min_weight
```

Now, let's run the Prim's algorithm on the example graph:
Keep in mind that this example uses a simple representation of a graph, but for more complex cases, you might want to use a library such as NetworkX to represent and manipulate graphs.

```python
mst, min_weight = prim_algorithm(graph)
print("Minimum Spanning Tree: ", mst)
print("Minimum Weight: ", min_weight)
```

Use same apporach for Kruskal's algorithm

```python
class UnionFind:
    def __init__(self, vertices):
        self.parent = {v: v for v in vertices}
        self.rank = {v: 0 for v in vertices}

    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        return self.parent[i]

    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)

        if root_i != root_j:
            if self.rank[root_i] < self.rank[root_j]:
                self.parent[root_i] = root_j
            elif self.rank[root_i] > self.rank[root_j]:
                self.parent[root_j] = root_i
            else:
                self.parent[root_j] = root_i
                self.rank[root_i] += 1
            return True
        return False

def kruskal_algorithm(graph):
    min_tree = []
    edges = []
    vertices = []

    # Extract all edges and sort them by weight
    for u in graph:
        vertices.append(u)
        for v, weight in graph[u].items():
            # Add each edge only once (e.g., from u to v, but not v to u again)
            if (weight, u, v) not in edges and (weight, v, u) not in edges:
                edges.append((weight, u, v))
    edges.sort()

    # Initialize Union-Find structure
    uf = UnionFind(vertices)
    min_weight = 0

    for weight, u, v in edges:
        if uf.union(u, v):
            min_tree.append((u, v, weight))
            min_weight += weight

    return min_tree, min_weight
```

```python
mst_kruskal, min_weight_kruskal = kruskal_algorithm(sample_graph)
print("Minimum Spanning Tree for sample_graph (Kruskal's): ", mst_kruskal)
print("Minimum Weight for sample_graph (Kruskal's): ", min_weight_kruskal)
```
